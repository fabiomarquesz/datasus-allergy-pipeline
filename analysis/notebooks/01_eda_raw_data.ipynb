###  As 10 principais causas de internação em SP (01/2023)

```python
top_doencas = df['diag_princ'].value_counts().head(10).reset_index()
top_doencas.columns = ['codigo_cid', 'total_internacoes']

print("--- Top 10 Causas de Internação (Geral) ---")
display(top_doencas)
```

#### Com base na Classificação Internacional de Doenças (CID-10), aqui está o significado de cada código listado na tabela:

#### O800 (O80.0): Parto normal (espontâneo).

#### O809 (O80.9): Parto único espontâneo, não especificado.

#### J189 (J18.9): Pneumonia, não especificada.

#### I64: Acidente vascular cerebral (AVC), não especificado como hemorrágico ou isquêmico.

#### N390 (N39.0): Infecção do trato urinário (ITU) de localização não especificada.

#### Z302 (Z30.2): Esterilização (frequentemente associado a procedimentos como laqueadura ou vasectomia).

#### I219 (I21.9): Infarto agudo do miocárdio (ataque cardíaco) não especificado.

#### I500 (I50.0): Insuficiência cardíaca congestiva.

#### A419 (A41.9): Septicemia (infecção generalizada grave) não especificada.

#### J180 (J18.0): Broncopneumonia não especificada. 


### Foco do Projeto: Alergias e Asma

```python

# =====================================================================
# 2. O Nosso Foco: Filtrando os casos de Alergia / Asma
# =====================================================================
# J30: Rinite / L20: Dermatite atópica / T78: Reações alérgicas / J45: Asma
cids_alergia = ('J30', 'L20', 'T78', 'J45')

# Usamos 'startswith' para capturar tanto o CID geral (J45) quanto os subtipos (J450, J451, etc.)
# Convertendo para string antes com .astype(str) por segurança
mascara_alergias = df['diag_princ'].astype(str).str.startswith(cids_alergia)

df_alergias = df[mascara_alergias]

print(f"Total de internações encontradas com os nossos CIDs: {len(df_alergias)}\n")

# Agrupando os casos de alergia encontrados
alergias_agrupadas = df_alergias['diag_princ'].value_counts().reset_index()
alergias_agrupadas.columns = ['codigo_cid_alergia', 'total_internacoes']
display(alergias_agrupadas)
```
